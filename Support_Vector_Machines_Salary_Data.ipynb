```python
# converting the non numeric data into numeric using one hot encoding
salary_train1 = salary_train.iloc[:,0:13]
salary_train1 = pd.get_dummies(salary_train1)
salary_train1
```

```python
salary_test1 = salary_test.iloc[:,0:13]
salary_test1 = pd.get_dummies(salary_test1)
salary_test1
```

### Since the number of columns are more, we'll use PCA (feature engineering) to decide the best features

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
salary_train_scaler = pd.DataFrame(trans.fit_transform(salary_train1))
salary_test_scaler = pd.DataFrame(trans.fit_transform(salary_test1))
```

```python
pca_train = PCA(n_components = 102)
pca_train_values = pca_train.fit_transform(salary_train_scaler)
var = pca_train.explained_variance_ratio_
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca_test = PCA(n_components = 102)
pca_test_values = pca_test.fit_transform(salary_test_scaler)
```

```python
plt.figure(figsize=(10,5))
plt.plot(var1)
```

### Based on the results of PCA, selecting first 50 PCA's out of 102

```python
finaltrain = pd.concat([pd.DataFrame(pca_train_values[:,0:50]),
                     salary_train[['Salary']]], axis = 1)
finaltrain
```

```python
finaltest = pd.concat([pd.DataFrame(pca_test_values[:,0:50]),
                     salary_test[['Salary']]], axis = 1)
finaltest
```

```python
# Using some part of data as comands are taking really long time to execute with such big data set
array = finaltrain.values
X = array[0:2000:,0:50]
Y = array[0:2000:,50]
```

### Building Classification Models

### KNN Classification Model

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Using the Grid search CV to find out best value for K
n_neighbors = np.array(range(1,40))
param_grid = dict(n_neighbors=n_neighbors)
model = KNeighborsClassifier()
grid = GridSearchCV(estimator=model, param_grid=param_grid)
grid.fit(X, Y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
# Visualizing the CV results
import matplotlib.pyplot as plt 
%matplotlib inline
# Choosing k between 1 to 41
k_range = range(1, 41)
k_scores = []
# Using the iteration to caclulator different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=5)
    k_scores.append(scores.mean())
# Plotting to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
# KNN Classification
num_folds = 10
kfold = KFold(n_splits=10)
model = KNeighborsClassifier(n_neighbors=29)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
# using some part of test data
array1 = finaltest.values
x = array1[0:2000:,0:50]
y = array1[0:2000:,50]
```

```python
clf = SVC(C= 15, gamma = 50)
clf.fit(x , y)
y_pred = clf.predict(x)
acc = accuracy_score(y, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y, y_pred)
```

### Using the various ensemble techniques to check the accuracy %

### 1. Bagging

```python
# Bagged Decision Trees for Classification
from sklearn.ensemble import BaggingClassifier
seed = 7
kfold = KFold(n_splits=15, random_state=seed, shuffle=True)
cart = DecisionTreeClassifier()
num_trees = 100
model = BaggingClassifier(base_estimator=cart, n_estimators=num_trees, random_state=seed)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```

### 2. Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
num_trees = 200
max_features = 4
kfold = KFold(n_splits=15)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```

### 3. Boosting

```python
# AdaBoost Classification
from sklearn.ensemble import AdaBoostClassifier
num_trees = 200
seed=7
kfold = KFold(n_splits=15, random_state=seed, shuffle=True)
model = AdaBoostClassifier(n_estimators=num_trees, random_state=seed)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```

### 4. Stacking

```python
# Stacking Ensemble for Classification
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
```

```python
# Creating the sub models
estimators = []
model1 = LogisticRegression(max_iter=500)
estimators.append(('logistic', model1))
model2 = DecisionTreeClassifier()
estimators.append(('cart', model2))
model3 = SVC()
estimators.append(('svm', model3))

# Creating the ensemble model
ensemble = VotingClassifier(estimators)
results = cross_val_score(ensemble, x, y, cv=kfold)
print(results.mean())
```

### Hence, we can conclude that, all the ensemble techniques provide almost same accuracy, but maximum accuracy is achieved through SVM classification method.
